```python
from google.colab import drive
import ipywidgets as widgets
from IPython.display import display, clear_output, HTML
import os , json
```

```python
# Mount Google Drive
drive.mount('/content/drive')
drive_path = '/content/drive/My Drive/Cloud Computing/'
os.chdir(drive_path)
```

```python
with open("Students.json","r") as file:
    students_data=json.load(file)
```

```python
# Dropdown for selecting a student
student_dropdown = widgets.Dropdown(
    options=[student['name'] for student in students_data['students']],
    description='Student:',
    style={'description_width': 'initial'},
    layout=widgets.Layout(width='50%'),
)

# Output area styled
output_area = widgets.Output(layout=widgets.Layout(border='1px solid green', padding='10px', width='60%'))

# Title for the form
title = widgets.HTML("<h2 style='color: #4CAF50;'>Student Information Form</h2>")

# Input field and button for adding a favorite movie
movie_input = widgets.Text(
    description='Favorite Movie:',
    placeholder='Enter favorite movie...',
    style={'description_width': 'initial'},
    layout=widgets.Layout(width='50%'),
)
add_movie_button = widgets.Button(
    description='Add Favorite Movie',
    button_style='success',
    icon='plus',
    layout=widgets.Layout(width='30%'),
)

# Function to display student info
def display_student_info(change):
    with output_area:
        clear_output(wait=True)
        selected_name = change['new']
        for student in students_data['students']:
            if student['name'] == selected_name:
                courses_list = "<ul>" + "".join(f"<li>{course}</li>" for course in student['courses']) + "</ul>"
                favorite_movie = student.get('movie', 'N/A')  # Default to 'N/A' if 'movie' is not present
                display(HTML(f"""
                <div style="font-family: Arial, sans-serif; color: #888;">
                    <h3 style="color: #2196F3;">Details for {selected_name}</h3>
                    <p><strong>Email:</strong> {student['email']}</p>
                    <p><strong>Courses:</strong> {courses_list}</p>
                    <p><strong>Favorite Movie:</strong> {favorite_movie}</p>
                    <p><strong>Link:</strong> <a href="{student['link']}" target="_blank">View Link</a></p>
                </div>
                """))

# Function to update the favorite movie in the data and file
def add_favorite_movie(b):
    selected_name = student_dropdown.value
    new_movie = movie_input.value.strip()
    for student in students_data['students']:
        if student['name'] == selected_name:
            student['movie'] = new_movie
            break

    # Save updated data to JSON
    with open(drive_path, "w") as file:
        json.dump(students_data, file, indent=4)

    with output_area:
        clear_output(wait=True)
        display(HTML(f"<p style='color: green;'>Successfully added '{new_movie}' as the favorite movie for {selected_name}.</p>"))

# Observe dropdown changes
student_dropdown.observe(display_student_info, names='value')

# Attach the button to the function
add_movie_button.on_click(add_favorite_movie)

# Display the UI
display(widgets.VBox([title, student_dropdown, movie_input, add_movie_button, output_area]))
```

```python
# Dropdown for selecting a student
student_dropdown = widgets.Dropdown(
    options=[student['name'] for student in students_data['students']],
    description='Student:',
    style={'description_width': 'initial'},
    layout=widgets.Layout(width='50%'),
)

# Input field and button for updating a favorite movie
favorite_movie_input = widgets.Text(
    placeholder="Enter favorite movie...",
    description="Favorite Movie:",
    style={'description_width': 'initial'},
    layout=widgets.Layout(width='50%'),
)

update_button = widgets.Button(
    description="Add Favorite Movie",
    style=widgets.ButtonStyle(
        button_color='green',
        font_weight='bold',
        tooltip='Click me to update!',
    ),
)

# Output area to display messages or results
output_area = widgets.Output(layout=widgets.Layout(border='1px solid green', padding='10px', width='60%'))

# Title for the form
title = widgets.HTML("<h2 style='color: #4CAF50;'>Update Student Information</h2>")


def display_student_info(change):
    with output_area:
        clear_output(wait=True)
        selected_name = change['new']
        for student in students_data['students']:
            if student['name'] == selected_name:
                courses_list = "<ul>" + "".join(f"<li>{course}</li>" for course in student['courses']) + "</ul>"
                favorite_movie = student.get('Favorite Movie', 'N/A')   # Get the favorite movie or display "N/A" if not available
                display(HTML(f"""
                <div style="font-family: Arial, sans-serif; color: #888;">
                    <h3 style="color: #2196F3;">Details for {selected_name}</h3>
                    <p><strong>Email:</strong> {student['email']}</p>
                    <p><strong>Courses:</strong> {courses_list}</p>
                    <p><strong>Favorite Movie:</strong> {favorite_movie}</p>
                    <p><strong>Link:</strong> <a href="{student['link']}" target="_blank">View Link</a></p>
                </div>
                """))

# Function to add or update a student's favorite movie
def add_favorite_movie(b):
    selected_name = student_dropdown.value
    new_favorite_movie = favorite_movie_input.value.strip()
     # Validate input: ensure both a student is selected and a movie is entered
    if not selected_name or not new_favorite_movie:
        with output_area:
            clear_output(wait=True)
            print("Please select a student and enter a valid movie name.")
        return
    # Update the student's favorite movie in the data
    for student in students_data['students']:
        if student['name'] == selected_name:
            student['Favorite Movie'] = new_favorite_movie
    with open(drive_path, "w") as file: # Save updated data back to the JSON file
        json.dump(students_data, file, indent=4)
    with output_area:
        clear_output(wait=True)
        print(f"Updated {selected_name}'s favorite movie to '{new_favorite_movie}'.")


student_dropdown.observe(display_student_info, names='value')
update_button.on_click(add_favorite_movie)

# Display the UI
display(widgets.VBox([student_dropdown, favorite_movie_input, update_button, output_area]))
```
